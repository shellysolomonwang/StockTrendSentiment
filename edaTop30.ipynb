```python
top_30_df = pd.read_excel('./data/top_30_summary_final.xlsx', index_col =None).drop(columns=['Unnamed: 0'])
```

# Compare the trend of `comment_counts`	`post_counts`	`total_counts`

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python

# loading libraries
import pandas as pd
import matplotlib.pyplot as plt


# Setting the positions and width for the bars
pos = list(range(len(top_30_df)))
width = 0.25

# Plotting the bars
fig, ax = plt.subplots(figsize=(15,6))

plt.bar(pos, top_30_df.total_counts, width, alpha=0.5, color='#EE3224')
plt.bar([p + width for p in pos], top_30_df.comment_counts, width, alpha=0.5, color='#F78F1E')
plt.bar([p + width*2 for p in pos], top_30_df.post_counts, width, alpha=0.5, color='#FFC222')

# Setting the y axis label
ax.set_ylabel('No. of comments/posts mentioned since 2018-01-01')

# Setting the chart's title
ax.set_title('Top 30 Stocks mentioned in r/wallstreetbets')

# Setting the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Setting the labels for the x ticks
ax.set_xticklabels(top_30_df.index)
plt.xticks(rotation=50)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, top_30_df.total_counts.max()+500] )

# Adding the legend and showing the plot
plt.legend([ 'Total','Daily Discussion Comments', 'Other Posts'], loc='upper right')
plt.grid()
plt.show()
```

```python

# loading libraries
import pandas as pd
import matplotlib.pyplot as plt


# Setting the positions and width for the bars
pos = list(range(len(top_30_df[:15])))
width = 0.25

# Plotting the bars
fig, ax = plt.subplots(figsize=(15,6))

plt.bar(pos, top_30_df.total_counts[:15], width, alpha=0.5, color='#EE3224')
plt.bar([p + width for p in pos], top_30_df.comment_counts[:15], width, alpha=0.5, color='#F78F1E')
plt.bar([p + width*2 for p in pos], top_30_df.post_counts[:15], width, alpha=0.5, color='#FFC222')

# Setting the y axis label
ax.set_ylabel('No. of comments/posts mentioned since 2018-01-01')

# Setting the chart's title
ax.set_title('Top 15 Stocks mentioned in r/wallstreetbets')

# Setting the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])

# Setting the labels for the x ticks
ax.set_xticklabels(top_30_df.index)
plt.xticks(rotation=50)

# Setting the x-axis and y-axis limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, top_30_df.total_counts.max()+500] )

# Adding the legend and showing the plot
plt.legend(['Total','Daily Discussion Comments', 'Other Posts'], loc='upper right')
plt.grid()
plt.show()
```
